```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
import joblib
```

# Step 1 - raining model

```python
df = pd.read_csv("../datasets/news_releases_tagged.csv")
df_copy = df.copy()

df_copy
```

```python
# features and labels
X = df_copy['clean']
y = df_copy['Tag']

# Ensure no NaN values in the text column
X = X.fillna('')
# Ensure labels are strings
y = y.astype(str)
# Check for any NaN values in the labels
if y.isnull().any():
    raise ValueError("Labels contain NaN values. Please clean the data.")
# Check for empty strings in the text column
if (X == '').any():
    raise ValueError("Text column contains empty strings. Please clean the data.")  

# TF-IDF vectorizer
vectorizer = TfidfVectorizer(max_features=5000)
X_vec = vectorizer.fit_transform(X)

# Split
X_train, X_test, y_train, y_test = train_test_split(X_vec, y, test_size=0.2, random_state=42, class_weight="balanced")

# Model training
model = LogisticRegression(max_iter=500)
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test) 

# Evaluation
print(classification_report(y_test, y_pred))

# Save the model and vectorizer
joblib.dump(model, '../models/news_classifier.pkl')
joblib.dump(vectorizer, '../models/tfidf_vectorizer.pkl')
```
